# Filtrado con `iloc` y `loc` en `Pandas`.

```python
import pandas as pd
```

```python
datos = pd.read_csv('../Datos/bestsellers.csv')
```

Si queremos filtrar las primeras cuatro filas:

```python
datos[0:4]
```

O queremos ver el vector asociado a la columna `Name`

```python
datos['Name']
```

O de forma conjunta a varias columnas:

```python
datos[['Name', 'Author']]
```

Con la función `loc` podemos filtrar unas filas, donde los extremos indicados son incluyentes.

```python
datos.loc[0:4]
```

Nótese que no se detiene en fila tres, sino que incluye a la fila cuatro. Ahora, si quiero especificar las columnas que me interesan:

```python
datos.loc[0:4, ['Name', 'Author']]
```

Y para poder ver los valores de una columna pero de forma tabular:

```python
datos.loc[:, ['Author']]
```

## Función `iloc`.

Esta función se diferencia con `loc` en que permite asignarle números a las columnas, de manera que para filtrar columnas o filas no tenemos que indicar de forma explícita el nombre que tenga, sino que basta con darle su índice.

```python
datos.iloc[:, 0:3]
```
