# Filtrado por condiciones con `Pandas`.

```python
import pandas as pd
import numpy as np
```

```python
datos = pd.read_csv('../Datos/bestsellers.csv')
datos.head()
```

Filtremos las observaciones que cumplen que el año sea mayor a 2016.

```python
datos['Year'] > 2016
```

Nótese que el retorno es una lista *booleana*. Para obtener la tabla, podemos pasarla como un *slicing* para el objeto donde tenemos los datos:

```python
datos[datos['Year'] > 2016]
```

Nótese que se devuelven 150 registros únicamente, que son todas aquellas observaciones que cumplen con la condición de que el año asociado es mayor estrictamente a 2016.

Filtremos ahora los libros que sean de ficción y que su fecha de publicación sea del 2019 en adelante.

```python
datos[(datos['Genre'] == 'Fiction') & (datos['Year'] > 2018)]
```
